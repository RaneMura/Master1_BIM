```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,r):
    return r*(1-x)*x

t = np.arange(0.0,1.0,0.01)

liste = [2.0,3.05,3.45,4.0]
for i in liste:
    plt.plot(t,f(t,i),label = "f(x)")
    plt.plot(t,t,label = "x")
    plt.title("Reprsentation de f(x) avec r : "+str(i)+"")
    plt.xlabel("x(p)")
    plt.ylabel("x(p+1)")
    plt.legend()
    plt.savefig("fx_"+str(i)+".pdf")
    plt.show()
```

```python
def logistic4(x1,x2,r,n):

    xValues1 = [x1]
    xValues2 = [x2]
    pValues = [i for i in range(0,n+1)]
    for j in pValues:
        if j==0:
            xValues1.append(x1)
            xValues2.append(x2)
        if j>0 : 
            xValues1.append(f(xValues1[j],r))
            xValues2.append(f(xValues2[j],r))
    xv1 =  np.array(xValues1[1::])      
    xv2 = np.array(xValues2[1::])
    diff = np.log(abs(xv2-xv1))
    return pValues,diff
```

```python
pv1,ldif = logistic4(0.01,0.00999999,3.8,100)

plt.plot(pv1,ldif,label = "diffrence")
plt.title("Reprsentation de xp en semi-log en fonction de p avec r : "+str(3.8)+"")
plt.xlabel("p")
plt.ylabel("x(p)")
plt.legend()
plt.savefig("log42.png")
plt.show()
```
